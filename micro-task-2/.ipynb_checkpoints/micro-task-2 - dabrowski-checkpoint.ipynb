```python
# Importing modules and libraries
from datetime import datetime
from elasticsearch import Elasticsearch
from elasticsearch_dsl import Search
from pprint import pprint
from perceval.backends.core.github import GitHub
import pandas as pd
import subprocess
import matplotlib.pyplot as plt
```

```python
# Instate ES and indicate repot to be used for further analysis, grimoirelab/perceval was selected for repository analysis
es = Elasticsearch('http://localhost:9200')
token = 'xyz'
```

```python
# Running the p2o.py to extract repository data and populate index and enriched_index in ES
subprocess.run(['p2o.py', '--enrich', '--index', 'github_raw',
      '--index-enrich', 'github', '-e', 'http://localhost:9200',
      '--no_inc', '--debug', 'github', 'grimoirelab', 'perceval',
      '-t', token, '--sleep-for-rate'])
```

```python
# Building a query definition
s = Search(using=es, index='github')
# Query only the data we need for the analysis
s = s.source(["item_type", "time_open_days", "time_to_close_days"])
# Filter only isssue created in the last six months and order in accordance with date the issues were created
s = s.filter('terms', item_type=['issue'])
s = s.filter('range', created_at={'gte':'now-6M'})
s = s.sort({'created_at':{'order':'asc'}})
# Keep the amount of 10000 results
s = s[0:10000]
```

```python
# Execute the query
result = s.execute()
# Convert results to dictionary
result = result.to_dict()['hits']['hits']

# List comprehension based on the issues from hits
issue_list = [item['_source'] for item in result]

# Create a dataframe for the issue
issue_list = pd.DataFrame(issue_list)

# Relace NaN values with zeros and rounding time_open_days and time_to_close_days
issue_list = issue_list.fillna(0)
issue_list = issue_list.round(0)
```

```python
# Aggregate open/closed issues in accordance to open/close days and count the number of issue in each bucket  
by_number_of_days_closed = issue_list.groupby([issue_list.time_to_close_days]).agg('count')
by_number_of_days_open = issue_list.groupby([issue_list.time_open_days]).agg('count')

```

```python
# Remove dublicate columns and merge together as a one dataframe
by_number_of_days_closed = by_number_of_days_closed.ix[:,[1]]
by_number_of_days_open = by_number_of_days_open.ix[:,[1]]
```

```python
# Plot distribution of time-to-close for closed issues x-> days, y-> number of closed issues
plot1 = by_number_of_days_closed.plot.bar(title='Distribution of time-to-close for closed issues', 
                                          legend=False, figsize=(10,5)) 
plot1.set_ylabel("Number of issues")
plot1.set_xlabel("Days")
```

```python
# Plot distribution of time-to-close for closed issues x-> days, y-> number of closed issues
plot1 = by_number_of_days_open.plot.bar(title='Distribution of time-to-close for opened issues', 
                                          legend=False, figsize=(10,5)) 
plot1.set_ylabel("Number of issues")
plot1.set_xlabel("Days")
```

```python
# Combine two dataframes and plot the picture for both closed and opened issues
result_dataframe = pd.concat([by_number_of_days_closed, by_number_of_days_open], axis=1, join='inner')
plot2 = result_dataframe.plot.bar(title='Distribution of time-to-close for opened issues and closed', 
                                          legend=True, figsize=(17,7), fontsize=15);
plot2.set_ylabel("Number of issues", fontsize=15)
plot2.set_xlabel("Days", fontsize=15)
plot2.legend(["Closed issues", "Opened issues"], fontsize=15);
```
